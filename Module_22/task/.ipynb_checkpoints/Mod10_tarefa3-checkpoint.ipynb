1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Carregar os dados
df = pd.read_csv("previsao_de_renda.csv")

# Exibir informações básicas
df.info()
````

````python
df.head()
````

````python
# Transformar em datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])

#Verificar transformação
df.info()
````

````python
# Variável qualitativa, por exemplo 'tipo_renda'
coluna_categ = 'tipo_renda'

# Agrupar por mês e categoria
dist = df.groupby([df['data_ref'].dt.to_period('M'), coluna_categ]).size().unstack()
````

````python
# Plotar gráfico de barras empilhadas
dist.plot(kind="bar", stacked=True, figsize=(12,6))

plt.title(f"Distribuição de {coluna_categ} ao longo do tempo")
plt.xlabel("Tempo (mensal)")
plt.ylabel("Quantidade")
plt.xticks(rotation=45)
plt.legend(title=coluna_categ)
plt.show()
````

````python
df[coluna_categ].value_counts()
````

### Conclusão da análise:
#### De modo geral, não foram observadas mudanças significativas na distribuição do **tipo de renda** ao longo do período analisado, o que indica uma relativa estabilidade dessa variável.  
#### Uma observação pontual é em relação ao tipo de renda **"Bolsista"**, que apresentou apenas quatro registros concentrados entre **02/2015 e 03/2015**, sugerindo tratar-se de casos isolados e não de um padrão recorrente.


2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# Selecionar variáveis qualitativas
qualitativas = df.select_dtypes(include=['object', 'category']).columns.tolist()
````

````python
qualitativas
````

````python

# Loop para gerar gráficos de barras empilhadas
for col in qualitativas:
    # Agrupar por tempo e categoria, calcular proporção
    dist = (df.groupby([df['data_ref'].dt.to_period('M'), col])
              .size()
              .groupby(level=0).apply(lambda x: x / x.sum())
              .unstack(fill_value=0))

    # Plotar gráfico
    dist.plot(kind="bar", stacked=True, figsize=(12,6))
    plt.title(f"Distribuição proporcional de '{col}' ao longo do tempo")
    plt.xlabel("Tempo (mensal)")
    plt.ylabel("Proporção")
    plt.xticks(rotation=45)
    plt.legend(title=col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
````

### Conclusão da análise:
#### De forma geral, não foram observadas alterações relevantes na distribuição das variáveis **sexo**, **tipo_renda**, **estado_civil** e **tipo_residencia**, que se mantiveram estáveis ao longo do período analisado.  
#### A variável **educação**, por outro lado, apresentou uma mudança importante: a partir de **08/2015** houve um aumento na proporção de pessoas com nível **superior completo**, enquanto a categoria **secundário** reduziu. Após esse ponto, a distribuição permaneceu estável até o fim da amostra.  
#### Considerando essa análise, pode-se concluir que as variáveis qualitativas analisadas são, em sua maioria, estáveis e, portanto, podem ser utilizadas em uma modelagem preditiva sem risco de introduzir forte viés temporal. Apenas a variável **educação** merece atenção, pois a mudança observada indica que seu impacto pode variar conforme o período analisado.
