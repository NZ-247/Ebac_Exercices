```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('./sinasc_RO_2019.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Gráfico de barras para a variável APGAR1
plt.figure(figsize=(8, 6))
df['APGAR1'].value_counts().plot(kind='bar')
plt.title('Distribuição da Pontuação APGAR1')
plt.xlabel('Pontuação APGAR1')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de baras para a variável APGAR5
plt.figure(figsize=(8, 6))
df['APGAR5'].value_counts().plot(kind='bar')
plt.title('Distribuição da Pontuação APGAR5')
plt.xlabel('Pontuação APGAR5')
plt.ylabel('Frequência')
```

```python
# Gráfico de barras para a variável SEXO
plt.figure(figsize=(8, 6))
df['SEXO'].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.title('Frequência por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de barras para a variável GESTACAO
plt.figure(figsize=(8, 6))
df['GESTACAO'].value_counts().plot(kind='bar')
plt.title('Distribuição da Duração da Gestação')
plt.xlabel('Semanas de Gestação')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de barras para variável ESTCIVMAE
cores = ['blue', 'green', 'yellow', 'purple', 'gray', 'red']
plt.figure(figsize=(8,6))
df['ESTCIVMAE'].value_counts().plot(kind='bar', color=cores)
plt.title('Distribuição do Estado Civil da Mãe')
plt.xlabel('Estado Civil da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Agrupando os dados por número de consultas e calculando a média do APGAR5
df_agrupado = df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# gráfico de barras
sns.barplot(x='CONSULTAS', y='APGAR5', data=df_agrupado)
plt.title('APGAR5 Médio por Número de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Criando o grid de gráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plotando os gráficos de contagem para cada coluna
sns.countplot(x='SEXO', data=df, ax=axes[0, 0])
sns.countplot(x='CONSULTAS', data=df, ax=axes[0, 1])
sns.countplot(x='ESTCIVMAE', data=df, ax=axes[1, 0])
sns.countplot(x='GESTACAO', data=df, ax=axes[1, 1])

# Ajustando os títulos dos gráficos
axes[0, 0].set_title('Frequência por Sexo')
axes[0, 1].set_title('Frequência por Número de Cosultas')
axes[1, 0].set_title('Frequência por Estado Civil da Mãe')
axes[1, 1].set_title('Frequência por Gestação')

# Rotacionando os rótulos do eixo x em 45 graus
plt.xticks(rotation=45)

# Ajuste dos rótulos dos eixos
for ax in axes.flat:
    ax.set_xlabel('')
    ax.set_ylabel('Frequência')

# Ajuste do espaçamento entre os gráficos
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```
