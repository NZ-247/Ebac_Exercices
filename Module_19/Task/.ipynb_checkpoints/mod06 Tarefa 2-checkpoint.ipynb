```python
# Carregando bibliotecas.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando base dados.

df = pd.read_csv('./sinasc_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Filtrando apenas os dados da coluna 'IDADEMAE'
data = df['IDADEMAE']

# Criando boxplot com a idade das Mães.
sns.boxplot(data=data)
plt.title('Coluna Idade Mãe')
plt.show
```

####  2) Faça um boxplot do peso do bebê

```python
df[['PESO']].boxplot()
```

####  3) Faça um histograma da idade da mãe

```python
df['IDADEMAE'].plot.hist(bins=20)
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data = df, alpha = .25, x = 'PESO',
            bins = 20, element = 'step', kde = True)
```
