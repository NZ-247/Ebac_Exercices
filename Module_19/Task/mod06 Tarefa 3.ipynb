#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Função para classificar o APGAR5:
def class_apgar(apgar_score):
    if 0 <= apgar_score <= 3:
        return 'Asfixia grave'
    elif 4<= apgar_score <= 5:
        return 'Asfixia Moderada'
    elif 6 <= apgar_score <= 7:
        return 'Asfixia leve'
    
    else:
      return 'Normal'
```

```python
# Aplicando função à coluna 'APGAR5' e criando nova coluna com sua categorização
df['CAT_APGAR5'] = df['APGAR5'].apply(class_apgar)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
#Verificação das variáveis 
df[['APGAR5', 'GESTACAO', 'SEXO']]
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab_cruzada = pd.crosstab(df['GESTACAO'], df['APGAR5'])
print(tab_cruzada)
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

Porcentagem APGAR5 por calsse de gestação.

```python
tab_pct = tab_cruzada.div(tab_cruzada.sum(axis=1), axis=0)
print(tab_pct)
```
