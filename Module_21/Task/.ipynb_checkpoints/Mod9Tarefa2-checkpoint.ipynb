### 1. Crie 2 faixas de Latitude do município (munResLat) sendo uma acima e outra abaixo de -10.5 e aplique o groupby usando essas faixas como chave e realize operações de soma, media, minimo, maximo, mediana, desvio padrao, variancia pra pelo menos 2 variáveis numéricas ainda não utilizadas

Criando coluna com faixas de latitude maior e menor que -10.5

```python
sinasc['faixa_latitude'] = sinasc['munResLat'].apply(lambda x: 'abaixo de -10.5' if x < -10.5 else 'acima de -10.5')
```

Calculando soma, media, minimo, maximo, mediana, desvio padrao e variancia das colunas 'IDADEMAE' e 'QTDFILVIVO':

```python
def calcular_agregacoes(df, coluna_grupo, colunas_calculo):
    """
    Calcula agregações (soma, média, mínimo, máximo, mediana, desvio padrão, variância) para as colunas especificadas, agrupadas por uma variável.

    Parâmetros:
    - df (DataFrame): O DataFrame a ser utilizado.
    - coluna_grupo (str): Nome da coluna para realizar o agrupamento.
    - colunas_calculo (list): Lista de colunas numéricas para calcular as métricas.

    Retorno:
    - DataFrame com os resultados das agregações.
    """
    return df.groupby(coluna_grupo, observed=True).agg({
        col: ['sum', 'mean', 'min', 'max', 'median', 'std', 'var'] for col in colunas_calculo
    })
```

```python
result_lat = calcular_agregacoes(df=sinasc, coluna_grupo='faixa_latitude', colunas_calculo=['IDADEMAE', 'QTDFILVIVO'])
```

```python
result_lat
```

```python
# Gráficos de dispersão entre IDH/PIB e os indicadores neonatais
def graficos_dispersao(df):
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    sns.scatterplot(ax=axes[0, 0], data=df, x='IDH_M', y='IDADEMAE', hue='faixa_IDH_M', palette='viridis')
    axes[0, 0].set_title('Relação entre IDH e Idade da mãe')

    sns.scatterplot(ax=axes[0, 1], data=df, x='PIB', y='IDADEMAE', hue='faixa_PIB', palette='coolwarm')
    axes[0, 1].set_title('Relação entre PIB e Idade da mãe')

    sns.scatterplot(ax=axes[1, 0], data=df, x='IDH_M', y='APGAR5', hue='faixa_IDH_M', palette='viridis')
    axes[1, 0].set_title('Relação entre IDH e APGAR5')

    sns.scatterplot(ax=axes[1, 1], data=df, x='PIB', y='APGAR5', hue='faixa_PIB', palette='coolwarm')
    axes[1, 1].set_title('Relação entre PIB e APGAR5')

    plt.tight_layout()
    plt.show()
```

```python
# Comparação de médias de peso, APGAR e frequência de anomalias por faixa_IDH_M e faixa_PIB
def comparar_medias(df):
    idh_comparacao = df.groupby('faixa_IDH_M', observed=False)[['PESO', 'APGAR5', 'IDANOMAL']].agg(['mean', 'std'])
    pib_comparacao = df.groupby('faixa_PIB', observed=False)[['PESO', 'APGAR5', 'IDANOMAL']].agg(['mean', 'std'])

    print("Médias e desvios padrão por faixa_IDH_M:")
    display(idh_comparacao)
    print("\nMédias e desvios padrão por faixa_PIB:")
    display(pib_comparacao)
```

```python
# Testar correlações entre IDH, PIB e indicadores neonatais
def testar_correlacoes(df):
    correlacoes = df[['IDH_M', 'PIB', 'PESO', 'APGAR5', 'IDANOMAL']].corr()
    print("\nMatriz de correlações:")
    display(correlacoes)

    # Heatmap de correlações
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlacoes, annot=True, cmap='coolwarm')
    plt.title('Heatmap de Correlações entre IDH, PIB e Indicadores Neonatais')
    plt.show()
```
