## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

### Resolução do execício 1.1):

````python
# 1) Seu código aqui
prova = {'mat':24, 'por':18, 'geo':8, 'ing':8, 'his':8, 'fis':12, 'qui':12}

qt_prova = pd.DataFrame(data=[prova])
qt_prova
````

````python
# lista com Data Frames das notas.
acertos = [df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui]

# Lista com os nomes das colunas (matérias)
colunas = ['mat', 'por', 'geo', 'ing', 'his', 'fis', 'qui']

#DataFrame vazio com o número de linhas igual ao de alunos (100 alunos)
df_notas = pd.DataFrame(index=range(100))

#Adicionando cada dataframe como coluna
for i, df in enumerate(acertos):
    df_notas[colunas[i]] = df['Qt_acertos']

df_notas
````

````python
# Criando dataframe com percentual de notas
df_prct = pd.DataFrame(index=df_notas.index)
for coluna in colunas:
    df_prct[coluna] = (df_notas[coluna] / qt_prova[coluna].values[0]*100).round(2)

# Df com percentual de acertos dos alunos por matéria
df_prct
````

### Resolunção do exercício 1.2):

````python
# 2) Seu código aqui
# Somando os acertos de todas na coluna total 
df_notas['total'] = df_notas.sum(axis=1)
df_notas
````

### Resolunção do exercício 1.3):

````python
# 3) Seu código aqui
#Total de questões da prova
tot_qt = qt_prova.sum(axis=1)

#Adicionando coluna de % de acertos dividindo o total de acertos pelo total de questões. 
df_notas['%_acertos'] = ((df_notas['total'] / tot_qt[0])*100).round(2)
df_notas
````

### Resolunção do exercício 1.4):

````python
# 4) Seu código aqui
#adicina nova coluna 'aprovado', que recebe 1 se a nota for superior à 45 e 0 para notas abaixo disso.
df_notas['aprovado'] = (df_notas['total'] > 45).astype(int)
df_notas
````

````python
#Calcula total de alunos com mais de 45 acertos
aprovados = df_notas['aprovado'].sum()
print(f'Apenas {aprovados} alunos alcançaram 45 acertos ou mais')
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

### Resolunção do exercício 2.1):

````python
# seu código aqui
#DataFrame com as colunas 'mat', 'fis' e 'qui'
df_eliminados = df_notas[['mat', 'fis', 'qui']]
# Vorifica se a linha tem o valor 0
df_eliminados = df_eliminados.apply(lambda row: any(row == 0), axis=1)
#Criando um DataFrame com os alunos desqualificados
df_eliminados = df_notas[df_eliminados]
df_eliminados
````

### Resolunção do exercício 2.2):

````python
qt_eliminado = df_eliminados.shape[0]
print(f'{qt_eliminado} alunos foram eliminados no vestibular por "zerar" em Matemática, física ou química.')
````

### Resolunção do exercício 2.3):

````python
# calculando média das notas dos alunos eliminados em geografia e história 
df_eliminados[['his', 'geo']].mean()
````

### Resolução do exercício 2.4):

````python
#  DataFrame com os alunos que obtiveram 45 acertos ou mais na prova e não "Zeraram" em matematica, física e química
#  passaram para a segunda fase do vestibular.
df_2fase = df_notas[(df_notas['aprovado'] == 1) & ~(df_notas[['mat', 'fis', 'qui']] == 0).any(axis=1)]

df_2fase
````
