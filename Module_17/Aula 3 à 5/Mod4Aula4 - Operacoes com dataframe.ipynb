# Registros de Vacinação COVID19 (Acre)

link: https://opendatasus.saude.gov.br/dataset/covid-19-vacinacao/resource/ef3bd0b8-b605-474b-9ae5-c97390c197a8

## Lendo CSV

```python
gripeMT = pd.read_csv(
    'https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/SGL/2024/uf=MT/lote=1/part-00000-6a00081d-d2e5-499b-bfa9-6b34ea5c7ba4.c000.csv',
    sep=';')
```

```python
gripeMT
```

## Verificando dataframe

```python
gripeMT.info()
```

## Copiando parte do dataframe

```python
# vacAcre_copy = vacAcre[:20000].copy()
# vacAcre_copy
gripeAF = gripeMT[gripeMT['municipio'] == 'Alta Floresta']
gripeAF
```

## Qual o percentual das vacinas aplicadas foram na capital???

```python
gripeMT['municipio'].unique()
```

```python
gripeCBA = gripeMT[gripeMT['municipio'] == 'Cuiabá']
gripeCBA
```

```python
gripeCBA.shape[0]
```

```python
percent = (gripeCBA.shape[0]/gripeMT.shape[0])*100
print(f'O Percentual de casos registrados na capital foi de {percent:.2f}%')
```

```python
# Contando os casos em cada município
casos_por_municipio = gripeMT['municipio'].value_counts()

# Calculando a proporção de cada município
proporcoes = casos_por_municipio / casos_por_municipio.sum()

# Criando um dicionário para agrupar os municípios com menos de 1%
agrupados = {'Menores que 1%': 0}
for municipio, proporcao in proporcoes.items():
    if proporcao < 0.01:
        agrupados['Menores que 1%'] += proporcao
    else:
        agrupados[municipio] = proporcao

# Convertendo o dicionário para uma Series
agrupados = pd.Series(agrupados)
```

```python
# Criando o gráfico de pizza
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(20, 18))
plt.pie(agrupados.values, labels=agrupados.index, autopct='%1.1f%%', startangle=90)
plt.title('Casos de gripe por municípios em MT')
plt.axis('equal')
plt.show()
```

## Qual a idade média dos pacientes???
