Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

### Resoloção da atividade 2.1)

````python
def get_pvl(uf_busca, tipo_interessado_busca):
    """
    Realiza uma consulta à API do Tesouro Nacional para obter dados de PVI (Prestação de Contas de Valores Imobilizados) 
    para uma determinada UF e tipo de interessado.

    Args:
        uf_busca (str): Sigla da UF (em qualquer caso) para a qual se deseja buscar os dados.
        tipo_interessado_busca (str): Tipo de interessado ("Estado" ou "Município") para a consulta.

    Returns:
        pandas.DataFrame: DataFrame contendo os dados da consulta à API, caso a consulta seja bem-sucedida.
        None: Caso a UF ou o tipo de interessado sejam inválidos.
    """
    
    uf = uf_busca.upper()  # Converte a UF para maiúsculas
    estados = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SE', 'SP', 'TO']  # Lista de estados válidos
    tipos_interessado_list = ['ESTADO', 'MUNICÍPIO']  # Lista de tipos de interessado válidos
    tipo_interessado = tipo_interessado_busca.upper()  # Converte o tipo de interessado para maiúsculas

    try:
        # Verifica se a UF informada é válida
        if uf not in estados:
            raise ValueError(f'Informe uma UF válida como: {estados}')
        
        # Verifica se o tipo de interessado é válido
        if tipo_interessado not in tipos_interessado_list:
            raise ValueError('Informe o tipo do interessado ("Estado" ou "Município")!')
        
        # Monta a URL da API com base na UF e no tipo de interessado
        if tipo_interessado == 'ESTADO':
            url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Estado'
        elif tipo_interessado == 'MUNICÍPIO':
            url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Munic%C3%ADpio'

        # Realiza a requisição à API
        r = requests.get(url)

        # Verifica se a requisição foi bem-sucedida
        if r.status_code == 200:
            # Converte a resposta da API para um DataFrame Pandas
            data_uf = r.json()
            df = pd.DataFrame(data_uf['items'])

            # Retorna o DataFrame com os dados
            return df
        else:
            raise Exception(f'Erro na requisição à API: {r.status_code}')
    except ValueError as e:
        print(e)
    except Exception as e:
        print(e)
````

````python
uf = 'mt'
interessado = 'Município'

df_mt = get_pvl(uf_busca=uf, tipo_interessado_busca=interessado)
````
