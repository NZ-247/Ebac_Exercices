#### Marcação de bom e mau
O objetivo da modelagem é classificar o risco de inadimplência, ou como se diz no meio, o risco de *default*. Podemos fazer longas discussões sobre o conceito de *default* com base em estudos e exigências regulatórias, para efeitos deste estudo, um cliente em *default* é aquele que está em 60 dias de atraso ou mais. Então classificaremos os clientes como 'bons' e 'maus' assim:
- **Maus** pagadores: são aqueles que entraram em 'default' (atraso 60 dias ou mais) nos 24 meses seguintes à aquisição do cartão de crédito. 
- **Bons** pagadores: são considerados todos os demais.
- **Excluídos**: Clientes que não adquiriram um cartão de crédito (seja por recusa, seja por desistência) não possuem informações de pagamento, portanto não se pode identificar se são bons ou maus. Há uma longa discussão e literatura sobre *inferência de rejeitados* que está fora do escopo deste exercício.

#### Construindo a variável resposta
A base de pagamentos está em um formato de 'base larga'. Essa base possui informações de pagamentos do cliente mês a mês a partir do mês de aquisição do crédito (mês 0) até o vigésimo quarto mês após a aquisição do crédito (mês 24). Utilizaremos essa base para determinar se um proponente é considerado 'bom pagador' ou caso apresente atraso representativo, será considerado 'mau pagador'.

#### Base larga vs base longa
A base ser larga significa que há uma linha para cada cliente, e que as informações estarão nas colunas, em contraste com a 'base longa', em que haveria uma linha para cada combinação cliente/mês, uma coluna indicando o cliente, outra indicando o mês, e apenas uma coluna com a informação do atraso.

````python
def taxa_null(data):
  """Calcula a taxa de valores faltantes para cada coluna em um DataFrame.

  Args:
    data (pd.DataFrame): O DataFrame para o qual a taxa de valores faltantes será calculada.

  Returns:
    dict: Um dicionário onde as chaves são os nomes das colunas e os valores são as taxas 
          de valores faltantes correspondentes, formatadas com três casas decimais.
  """
  values = data.isna().sum()
  linhas = data.shape[0]
  percent_dict = {}  # Inicializa um dicionário vazio
  for coluna in values.index:  # Itera pelas colunas do DataFrame
    taxa = f'{(values[coluna]/linhas)*100:.3f}%'
    percent_dict[coluna] = taxa  # Adiciona a coluna como chave e a taxa como valor
  return percent_dict  # Retorna o dicionário
````

````python
def fill_nulls(df):
    """
    Preenche valores faltantes em colunas com base no dado anterior e posterior não nulo de um DataFrame.

    Args:
        df (pd.DataFrame): DataFrame com colunas representando meses e 
                           linhas representando clientes.

    Returns:
        pd.DataFrame: DataFrame com os valores faltantes preenchidos.
    """

    # Itera pelas colunas de meses, pulando a coluna de ID e a primeira coluna de mês
    for i in range(2, len(df.columns) - 1):
        # Obtém os nomes das colunas dos meses atual, anterior e seguinte
        mes_atual = df.columns[i]
        mes_anterior = df.columns[i - 1]
        mes_seguinte = df.columns[i + 1]

        # Cria uma máscara para identificar os valores faltantes na coluna atual
        faltantes = df[mes_atual].isnull()

        # Preenche os valores faltantes com base nas condições
        df.loc[faltantes & (df[mes_anterior] == df[mes_seguinte]), mes_atual] = df[mes_anterior]

    # Trata a última coluna de mês separadamente
    ultima_coluna = df.columns[-1]
    penultima_coluna = df.columns[-2]
    faltantes_ultima = df[ultima_coluna].isnull()
    df.loc[faltantes_ultima & (df[penultima_coluna] == df[penultima_coluna]), ultima_coluna] = df[penultima_coluna]

    return df
````

#### Tarefa 1) Marcar *default* no mês
Faça uma indicadora de se o cliente está em *default* em cada uma das marcações (mes_00 a mes_24). Dica: você pode utilizar o método ```.isin()``` do Pandas. Consulte a [documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isin.html) caso necessário.


````python
cat_default = ['2', '3', '4', '5'] #Categorias que representam default

meses = [f'mes_{i}' for i in range(25)] # Cria um lista de 'mes_0' a 'mes_24'
meses_def = []

# Itera sobre cada coluna de mês e cria uma coluna indicadora de defalt
for mes in meses:
    meses_def.append(f'default_{mes}')
    pg_clean[f'default_{mes}'] = pg_clean[mes].isin(cat_default).astype(int)


# lista vazia para armazenar as colunas reorganizadas
colunas_reorganizadas = ['ID']  # Inclui a coluna 'ID' ou outras colunas iniciais que não sejam meses

# Itera sobre os meses e reorganiza as colunas para que o indicador de default fique ao lado de sua respectiva coluna
for mes in range(25):
    colunas_reorganizadas.append(f'mes_{mes}')
    colunas_reorganizadas.append(f'default_mes_{mes}')

# Reorganiza as colunas do DataFrame com base na nova ordem
pg_clean = pg_clean[colunas_reorganizadas]

pg_clean
````

#### Tarefa 2) 'bons' e 'maus' ao longo de todos os 24 meses de desempenho
Marque para cada cliente se ele teve pelo menos um episódio de *default* entre o mês 0 e o mês 24. Dica: o método ```sum()``` pode ajudar. Caso precise, consulte a [documentação](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sum.html) e procure pelo argumento ```axis```, você viu outros métodos que possuem esse argumento também. Tendo o número de meses em default de cada cliente, basta marcar ```True``` para todos aqueles que possuem pelo menos 1 mês em *default* e ```False``` para os demais.


````python
# Lista de colunas indicadoras de default
col_def = [f'default_mes_{i}' for i in range(25)]

# Soma o número de meses em default para cada cliente.
pg_clean.loc[:, 'num_meses_default'] = pg_clean[col_def].sum(axis=1)
````

````python
# Cria coluna que indica se o cliente cometeu o default em ao menos um mês, recebe True se sim e False caso contrário.
pg_clean['teve_default'] = pg_clean['num_meses_default']

pg_clean[['ID', 'teve_default']]
````

````python
# Certificação de que a coluna 'teve_default' está no formato booleano
pg_clean['teve_default'] = pg_clean['teve_default'].astype(bool)

# Configurações de estilo
sns.set(style="whitegrid")

# Contagem de valores de 'teve_default'
default_counts = pg_clean['teve_default'].value_counts()

# Gráfico de Barras
plt.figure(figsize=(10, 5))
sns.barplot(x=default_counts.index, y=default_counts.values, palette="viridis")
plt.title("Proporção de Clientes com Default")
plt.xlabel("Teve Default")
plt.ylabel("Número de Clientes")
plt.xticks([0, 1], ['Não', 'Sim'])  # Define os rótulos para False (0) e True (1)
plt.show()

# Ajuste dinâmico para os rótulos do gráfico de pizza
labels = ['Não' if not val else 'Sim' for val in default_counts.index]

# Gráfico de Pizza
plt.figure(figsize=(8, 8))
plt.pie(default_counts, labels=labels, autopct='%1.1f%%', colors=['#66c2a5', '#fc8d62'], startangle=90)
plt.title("Proporção de Default entre Clientes")
plt.axis('equal')  # Assegura que o gráfico de pizza seja um círculo
plt.show()
````

#### Tarefa 3) Marcando proponentes expostos ao risco de crédito
Marcando proponentes que se tornaram tomadores: lembre-se de que clientes que não adquiriram o cartão devem ser desconsiderados. A base de pagamentos possui apenas clientes que adquiriram cartão de crédito, então você pode selecionar somente os clientes da base de propostas que se encontram na base de pagamentos.

````python
# Criar uma coluna 'target' na base de propostas, inicialmente com valor 0
propostas_clean['target'] = 0

# Obtem a lista de IDs únicos dos clientes que estão na base de pagamentos
clientes_com_pagamento = pg_clean['ID'].unique()

# Marcar como 1 (tomadores) os clientes da base de propostas que estão na lista de clientes com pagamento
propostas_clean.loc[propostas_clean['ID'].isin(clientes_com_pagamento), 'target'] = 1

# Exibir a contagem de proponentes e tomadores
print(propostas_clean['target'].value_counts())
````

````python
# Calcula a contagem de proponentes aceitos e recusados
contagem_propostas = propostas_clean['target'].value_counts()

# rótulos para o gráfico
rotulos = ['Proponentes Rejeitados', 'Proponentes Aceitos']

# Cria o gráfico de pizza
plt.pie(contagem_propostas, labels=rotulos, autopct='%1.1f%%', startangle=90)

# título ao gráfico
plt.title('Proporção de Proponentes')

# Garante que o círculo seja um círculo perfeito
plt.axis('equal')

plt.show()
````

Faça uma junção das informações da base de propostas com a variável de *default* que você acabou de construir. Talvez você consiga realizar a tarefa 3 e tarefa 4 em uma única linha de código ;)
